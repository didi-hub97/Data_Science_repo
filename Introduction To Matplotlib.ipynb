Matplotlib supports plotting multiple charts(subplots) on a single figure, which is Matplotlib's term for the drawing of canvas.

```python
%matplotlib inline
#will lead to static images of your plot embedded in the notebook
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
#This yields an empty figure object (0 Axes).
fig = plt.figure(figsize = (12, 6), dpi = 200)
```

### Plotting in subplots

Let'suse numpy.linspace(...) to generate evenly spaced values on the x-axis, and then the numpy.square(...), 
numpy.sin(...), and numpy.cos(...) methods to generate corresponding values on the y axis.

We will use the ax1, ax2, and ax3 axes variables we got from adding subplots to plot these functions:


```python
import numpy as np
np.random.seed(0)

x  = np.linspace(0, 1, num = 20)
y1 = np.square(x)
ax1.plot(x, y1, color = "black", linestyle = "--")
```

```python
y2 = np.sin(x)
ax2.plot(x, y2, color = "black", linestyle = ":")
```

```python
y3 = np.cos(x)
ax3.plot(x, y3, color = "black", linestyle = "-.")
```

```python
fig
```

The sharex= parameter can be passed when creating subplots to specify that all the subplots should share the same x axis.

Let's demonstrate this functionality and plot the square, and then use the numpy.power(...) method to raise xto the power of 10 and plot them with the same x axis:

```python
fig, (ax1, ax2) = plt.subplots(2, figsize = (12, 6), sharex = True)
ax1.plot(x, y1, color = "black", linestyle = "--")

y2 = np.power(x, 10)
ax2.plot(x, y2, color = "black", linestyle = "-.")
```

The carts we generated are not self-explanatory yet-it is unclear what the units on the x-axis and the y-axis are, and what each chart represents. To improve the charts, we need to enrich them with colors, markers and linestyle, to enrich the axes with ticks, legends, and labels and provide selected data points' annotations.

### Plots: Color, Markers and Line Styles:

Colors, markers and line styles make charts easier to understand.

The code block that follows plot four different functions and uses the following parameters to modify the appearance:

* The color=  parameter is used to asssign colors. 
* The linewidth= parameter is used to change the width/thickness of the lines.
* The marker= parameter assigns different shapes to mark the data points. 
* The markersize= parameter changes the size of those markers.
* The alpha= parameter is used to modify the transparency.
* The drawstyle= parameter changes the default line connectivity to step connectivity between data points for one plot.


Below are plots demonstrating different colors, line styles, marker styles, transparency and size options.

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, figsize = (12, 12), 
                                        sharex = True)

x = np.linspace(0, 10, num  = 20)
y1 = np.exp(x)
y2 = x ** 3
y3 = np.sin(y2)
y4 = np.random.randn(20)

ax1.plot(x, y1, color = "black", linestyle = "--",
        linewidth = 5, marker = "x", markersize = 15)

ax2.plot(x, y2, color = "green", linestyle = "-.", 
        linewidth = 2, marker = "^", markersize = 10, alpha = 0.9)

ax3.plot(x, y3, color = "red", linestyle = ":", 
        marker = "*", markersize = 15, drawstyle = "steps")

ax4.plot(x, y4, color = "green", linestyle = "-",
         marker = "s", markersize = 15)
```

### Axes: Ticks, Labels, Legends

The  charts can be further improved by customizing the axes via ticks, limits, and labels.

The matplotlib.pyplot.xlim(...) method sets the range of values on the x-axis.

The matplotlib.pyplot.xticks(...) method specifies where the ticks show up on the x-axis

This modifies the x-axis to be within the specified limits and the ticks at the explicitly specified values.

```python
plt.xlim([8, 10.5])
plt.xticks([8, 8.42, 9.47, 10, 10.5])
plt.plot(x, y1, color = "black", linestyle = "--", marker = "o")
```

We can also change the scale of the axis to non-linear using the matplot.Axes.sety_scale(...) method.

```python
## Then output of this code block shows the difference in the scale of the y axis, which is now
#logarithmic, and the x-axis ticks have the specific tick labels.

fig, ax = plt.subplots(1, figsize = (12, 6))

ax.set_yscale("log") #set the y-axis scale

ax.set_xticks(x)

ax.set_xticklabels(list("ABCDEFGHIJKLMNOPQRST"))

ax.plot(x, y1, color = "black", linestyle = "--", marker = "o", 
       label = "y = exp(x)")
```

Logarithmic scales in the chart above are useful if the dataset covers a large range of values and/or if we want to communicate percentage change or multiplicative factors.

In the chart above, the matplotlib.Axes.set_title(...) method adds a title to the plot and the matplotlib.Axes.set_xlabel(...) method sets labels for the x and y axes.

The matplotlib.Axes.legend(...) method adds a legend, which makes the plot easier to interpret. The log= parameter specifies the location of the legend on the plot with loc="best", meaning matplotlib picks the best location automatically:

```python
ax.set_title("xtickslabel example")

ax.set_xlabel("x labels")

ax.set_ylabel("log scale y values")

ax.legend(loc = "best")

fig
```

### Matplotlib and Pandas DataFrame

The pandas library provides plotting capabilites for series and dataframe using matplotlib.

Lets create a pandas dataframe with the cont value containing continuous values that mimc prices and the Delta1 and Delta2 values to mimic price changes. the Cat value contains categorical data from five possibilities

```python
import pandas as pd
```

```python
df = pd.DataFrame(index = range(1000), 
              columns = ["Cont value", "Delta1 value", "Delta2 value", "Cat value"])
df["Cont value"] = np.random.randn(1000).cumsum()
df["Delta1 value"] = np.random.randn(1000)
df["Delta2 value"] = np.random.randn(1000)
df["Cat value"] = np.random.permutation(["Very High", "High", "Medium", "Low", "Very Low"]*200)
df["Delta1 discrete"] = pd.cut(df["Delta1 value"], labels = [-2, -1, 0, 1, 2], bins = 5).astype(np.int64)
df["Delta2 discrete"] = pd.cut(df["Delta2 value"], labels = [-2, -1, 0, 1, 2], bins = 5).astype(np.int64)

df
```

```python
### Line plots of a Dataframe column

We can plot "Cont value" in a line plot using the pandas.DataFrame.plot(...) method with  this kind= parameter.
```

```python
###Line charts are typically used for displaying time series.

df.plot(y = "Cont value", kind = "line", color = "black", linestyle = "-", figsize = (12, 6))
```

### Bar plots of a dataframe column

The pandas.Dataframe.plot(...) method can be used with the kind = "bar" parameter to build a chart.

PLS NOTE THE USAGE: Bar plots are most suitable for comparing the magnitude of categorical variables/values. Do not use it for continuous variables

```python
### Let's first group the dataframe by the "Cat value" value, and then the plot
#The Delta1 discrete values counts in a bar chart.
#This generates the following plot showing the frequency of (Cat value, Delta1 discrete) values pairs:
###Value_count is used for categorical values similar to xtabs in R which returns the value count and 
#groupby does the cross tabulation

df.groupby("Cat value")["Delta1 discrete"].value_counts().plot(kind = "bar",)

```
